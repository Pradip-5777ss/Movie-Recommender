## Process the IMDB-5000 movies dataset from Kaggle

Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Read data from IMDB-5000 Movies dataset

```python
# imdb_movies_data = pd.read_csv('../../Data/Datasets/IMDB-5000_movie_metadata.csv')
imdb_movies_data = pd.read_csv('../Data/Datasets/IMDB-5000_movie_metadata.csv')
```

Plot the data present in the dataset. Here we have movies list upto 2016

```python
imdb_movies_data.title_year.value_counts(dropna=False).sort_index().plot(kind='barh', figsize=(15,25))
plt.show()
```

From the whole dataset we filter these columns. Recomendation will be based on these features

```python
imdb_movies_data = imdb_movies_data.loc[:,['director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name', 'genres', 'movie_title']]
```

Show the first 10 row with these selected columns

```python
# imdb_movies_data.head(10)
imdb_movies_data
```

Replace the unknown data columns with Nan

```python
imdb_movies_data['actor_1_name'] = imdb_movies_data['actor_1_name'].replace(np.nan, 'unknown')
imdb_movies_data['actor_2_name'] = imdb_movies_data['actor_2_name'].replace(np.nan, 'unknown')
imdb_movies_data['actor_3_name'] = imdb_movies_data['actor_3_name'].replace(np.nan, 'unknown')
imdb_movies_data['director_name'] = imdb_movies_data['director_name'].replace(np.nan, 'unknown')
```

```python
imdb_movies_data
```

Convert the "movie_title" column data in lowercase

```python
imdb_movies_data['movie_title'] = imdb_movies_data['movie_title'].str.lower()
```

```python
imdb_movies_data
```

Replace the '|'(separetor) with ' '(space) in "genres" column

```python
imdb_movies_data['genres'] = imdb_movies_data['genres'].str.replace('|',' ')
```

```python
imdb_movies_data
```

Null terminating character at the end

```python
imdb_movies_data['movie_title'][2]
```

Removing the null terminating character at the end of the the "movie_title"

```python
imdb_movies_data['movie_title'] = imdb_movies_data['movie_title'].apply(lambda x : x[:-1])
```

```python
imdb_movies_data['movie_title'][2]
```

```python
imdb_movies_data.to_csv('../Data/ProcessedData/imdb-5000_preprocessData.csv', index=False)
```
